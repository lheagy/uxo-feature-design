```python
import discretize
from SimPEG import maps
import numpy as np
import matplotlib.pyplot as plt

import BTInvert
import uxo_utils

from matplotlib import rcParams
rcParams["font.size"] = 14
```

```python
survey = uxo_utils.create_survey_from_file(
    "../../uxo-utils/data-blacktusk/Testplot5F.h5", 
    True
)
```

```python
fig, ax = plt.subplots(1, 1)
line_inds = survey.line == survey.unique_lines[0]
ax.plot(survey.easting[line_inds], survey.northing[line_inds], '.', ms=0.2, alpha=0.1)
ax.plot(survey.easting[line_inds][:1000], survey.northing[line_inds][:1000], '.', ms=0.5)
# ax.set_aspect(1)
```

```python
fig, ax = plt.subplots(1, 1)
for line_id in survey.unique_lines:
    line_inds = survey.line == line_id
    ax.plot(survey.easting[line_inds], survey.northing[line_inds], '.', alpha=0.1, ms=0.2)
ax.set_aspect(1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 8))
for line_id in survey.unique_lines:
    line_inds = survey.line == line_id
    ax.plot(
        survey.rotated_x[line_inds], survey.rotated_y[line_inds], '.', alpha=0.1, ms=0.2
    )
ax.set_aspect(1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 8))
line_inds = survey.line == survey.unique_lines[0]
ax.plot(survey.rotated_x[line_inds], survey.rotated_y[line_inds], '.', ms=0.2, alpha=0.1)
ax.plot(survey.rotated_x[line_inds][:1000], survey.rotated_y[line_inds][:1000], '.', ms=0.5)
ax.set_aspect(1)
```

```python
fig, ax = plt.subplots(1, 1)
for line_id in survey.unique_lines:
    line_inds = survey.line == line_id
    ax.plot(
        survey.rotated_y[line_inds], (survey.rotated_yaw[line_inds])*180/np.pi, '.', alpha=0.1, ms=0.2
    )
```
